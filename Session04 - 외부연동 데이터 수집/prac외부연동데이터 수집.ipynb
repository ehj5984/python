```python
# 라이브러리 선언
import pandas as pd
from bs4 import BeautifulSoup
import requests 

# targeturl 정의
targetUrl = "https://www.farfetch.com/kr/shopping/women/loewe/bags-purses-1/items.aspx?view=90&category=136315"

resp = requests.get(url = targetUrl)
resp.encoding= "utf-8"

html = resp.text

htmlObj = BeautifulSoup(html, "html.parser")

# 제품명과 가격 전체 데이터 추출
allName = htmlObj.findAll(name ="p", attrs ={"itemprop":"name"})
allPrice = htmlObj.findAll(name ="span", attrs ={"data-testid":"price"})

productList = []
priceList = []

# for 문을 이용해 텍스트 데이터 추출 후 밀어넣기 
for i in range(0,len(allName)):
    eachName = allName[i]
    TagText = eachName.text
    productList.append(TagText)
    
    eachPrice = allPrice[i]
    PriceText = eachPrice.text
    priceList.append(PriceText)
    
    
# 컬럼 리스트 정의     
columnList=["제품명","가격"]
# 컬럼 값 넣기 
pd.DataFrame(zip(productList,priceList), columns = columnList)




# AllData로 데이터 프레임 정의
AllData = pd.DataFrame(zip(productList,priceList), columns = columnList)

AllData
```

```python
# 라이브러리 선언

import pandas as pd
from bs4 import BeautifulSoup
import requests 

# targetUrl 선언 및 접속

targetUrl = "https://www.timeanddate.com/holidays/south-korea"
resp = requests.get(url=targetUrl)
resp.encoding="utf-8"
html = resp.text
htmlObject = BeautifulSoup(html, "html.parser")


# date 수집 후 list로 변환
find_date = htmlObject.findAll(name = "th", attrs = {"class": "nw"})

find_dateList=[]

for j in range(0, len(find_date)):
    eachDate = find_date[j]
    Date = eachDate.text
    find_dateList.append(Date)

# date 제외한 데이터들 수집 후 list로 변환
otherData = htmlObject.findAll(name = "td")

find_typeList=[]
find_dayList =[]
find_nameList=[]

for i in range (0,len(otherData), +3):
    Day = otherData[i].text
    find_dayList.append(day)

    Name = otherData[i+1].text
    find_nameList.append(Name)
    
    Type = otherData[i+2].text
    find_typeList.append(Type)
    
# 리스트를 데이터프레임 화
pd.DataFrame(zip(find_dateList,find_dayList,find_nameList,find_typeList))
# 컬럼리스트 정의
columnList = ["Date", "","Name","Type"]
# 컬럼값 대입
result = pd.DataFrame(zip(find_dateList,find_dayList,find_nameList,find_typeList), columns = columnList)
```
