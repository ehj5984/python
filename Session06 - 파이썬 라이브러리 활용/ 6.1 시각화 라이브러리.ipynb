### 1. 라이브러리 선언 

```python
import pandas as pd

import matplotlib.pyplot as plt

%matplotlib inline 
```

### 2. 데이터 불러오기 

```python
Data = pd.read_csv("../dataset/feature_regression_example.csv")

feateuresData2015 = Data.loc[Data.YEAR== 2015]

feateuresData2015.head(0)

# 팝업 창 활용하여 차트 시연
xValue = feateuresData2015.WEEK 
yValue = feateuresData2015.QTY

plt.plot(xValue,yValue , "bo--", lw=1.5, label= "2015")
# X, Y 축 라벨
plt.xlabel("WEEK")
plt.ylabel("QTY")
plt.legend(loc = 0) # 라벨을 먼저 부여하고 표식을 나중에 써줘야함
```

### ../dataset/feature_regression_example.csv 2015년도 자료만 활용하여 
### x축: week 이중축으로 qty, pro_percent를 차트로 그리세요

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 

Data = pd.read_csv("../dataset/feature_regression_example.csv")

feateuresData2015 = Data.loc[Data.YEAR== 2015]

# 팝업 창 활용하여 차트 시연
xValue = feateuresData2015.WEEK 
yValue = feateuresData2015.QTY
y2Value = feateuresData2015.PRO_PERCENT
# 이중축 
fig = plt.figure(figsize = (14,4))
ax1 = fig.add_subplot(111)
ax1.plot(xValue,yValue , "bo--", lw=1.5, label= "2015")
ax1.grid(True)
ax1.legend(loc = 0)
ax1.set_xlabel('ax1_value')
ax1.set_ylabel('ax1_index')
plt.title('subplot')

ax2 = ax1.twinx()
ax2.plot(xValue,y2Value , "go--", lw=1.5, label= "201515")
ax2.legend(loc=4)
ax2.set_ylabel('ax2_value')

```
