## 2. 데이터 불러오기

```python
selloutData = \
    pd.read_csv("../dataset/kopo_channel_seasonality_new.csv")
```

```python
selloutTypeData = selloutData.astype({
    "REGIONID":str,
    "PRODUCT":str,
    "YEARWEEK":str,
    "QTY":float})
```

## 3. 데이터 통합

```python
## YEARWEEK에서 나눠야할 연도, 주차 길이 정의
yearAndWeek = 4

## YearAndWeek를 기준으로 데이터 나누기
selloutTypeData["YEAR"]=selloutTypeData.YEARWEEK.str[:yearAndWeek]
selloutTypeData["WEEK"]=selloutTypeData.YEARWEEK.str[yearAndWeek:]

## 구하고자 하는 Week 변수 지정
yearlength = 52

## yearlength를 기준으로 데이터 찾기
findSelloutData = selloutTypeData.loc[selloutTypeData.WEEK.astype(int)<=yearlength]

## 인덱스 재설정
refinedSelloutData=findSelloutData.reset_index(drop=True)

## 검산식
# refinedSelloutData.loc[refinedSelloutData.WEEK.astype(int)>yearlength]

refinedSelloutData
```

```python
# [지역, 상품, 연도 별 집계] sortedData 에서 지역, 상품, 연도 단위 판매량(QTY_NEW) 의 평균 연산 후 groupData 변수에 담으세요
groupKey =["REGIONID","PRODUCT","YEAR"]

groupData = sortedData.groupby(by=groupKey)["QTY"].agg(["mean"])


## sortedData.groupby(by=groupKey)["QTY"].describe()

groupData.rename(columns={"mean":"QTY_NEW"}, inplace=True)



groupData.reset_index(inplace =True)

groupData
```

## 고급연산 (이동집계함수)
### 데이터프레임명.rolling(window=5,center=False).mean()

```python
# 이동평균 활용 추세선 생성(구간 5, 중심기준)
sortedData["MA5"] = sortedData.QTY.rolling(window=5, 
                                         min_periods =1, 
                                        center= True).mean()
```

```python
sortedData["MA5_BASIC"] = sortedData.QTY.rolling(window=5, 
                                                 # min_period =1, 
                                                 center= True).mean()
```

```python
sortedData["MA5_BASIC_NO"] = sortedData.QTY.rolling(window=5, 
                                                 # min_period =1, 
                                                 center= False).mean()
```

```python
sortedData["EFFECT"]= sortedData.QTY / sortedData.MA5
```

```python
sortedData.YEAR.drop_duplicates()
```

```python
groupKey = ["REGIONID","PRODUCT","WEEK"]
```

```python
# 3개 연도의 지역, 상품, 주차별 평균 effect를 뽑아내고 싶다
ratioData = sortedData.groupby(groupKey)[["EFFECT"]].agg("mean").reset_index()
```
