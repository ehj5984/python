## 소수점 둘째 짜리 반올림

```python
#def 함수이름, 파라미터
def roundFunction(inValue):
    # 여기서 inValue는 함수를 개발하기 위한 디버깅 코드

    # 1. 입력받은 데이터를 반올림 자리수 만큼 곱해준다

    # inValue = 1.5667924528301887
    step1 = inValue * 100

    # 2. 곱해준 결과에 0.5를 더한 후 소수점 이하 버림

    step2 = int(step1 + 0.5)

    # 3. 100으로 나눠준다

    step3 = step2 / 100
    outValue = step3
    # 리턴을 해야 아웃밸류를 내보냄
    return outValue
```

```python
def roundOption(inValue,optionValue):
#     inValue =1.2225
#     optionValue =1
    step1 = inValue * 10**optionValue

   # 2. 곱해준 결과에 0.5를 더한 후 소수점 이하 버림
    step2 = int(step1 + 0.5)
   # 3. 10**optionValue로 나눠준다
    step3 = step2 / 10**optionValue      
    outValue = step3

    return outValue
```
